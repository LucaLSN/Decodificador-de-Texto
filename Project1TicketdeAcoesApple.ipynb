#Importando Dados com Python

### Uma ação (também conhecida como patrimônio líquido) é um título que representa a propriedade de uma fração de uma empresa. Isso dá ao proprietário das ações uma proporção dos ativos e lucros da empresa igual à quantidade de ações que possui.

- Utilizando o módulo `Ticker` podemos criar um objeto que nos permitirá acessar funções para extrair dados. Conforme código abaixo:

- Bibliotecas utilizadas:
1. *yfinance:* ela nos permite extrair dados de ações retornando dados em um dataframe do pandas.
2. *pandas:* permite explorar o dataframe

```python
import yfinance as yf
import pandas as pd
#importar os dados acionários da empresa Apple
apple = yf.Ticker("AAPL")
'''Usando o método history() podemosobter o
preço das ações durante um determinado período de tempo.'''
preco_acoes = apple.history(period="1y")
#Gerando dataframe do pandas com a função head;
preco_acoes.head()
preco_acoes.reset_index(inplace=True)
#Plot do data frame
preco_acoes.plot(x="Date", y="Open")
```
